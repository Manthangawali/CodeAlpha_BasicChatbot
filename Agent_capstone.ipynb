```python
"""
PROJECT: Gift Scout Agent (Capstone)
AUTHOR: [Manthan Gawali]
TRACK: Concierge Agents

DESCRIPTION:
This agent acts as a personal shopping assistant. It uses the Google Gemini API
to generate curated gift suggestions based on a recipient persona and a strict budget.

KEY CONCEPTS IMPLEMENTED:
1. System Prompting (Defining the specific role of the AI).
2. Parametric Inputs (Injecting user data into the prompt).
3. Error Handling (Fallback logic if the primary model fails).
"""

import google.generativeai as genai
import time
import os

api_key = "INSERT_YOUR_GOOGLE_API_KEY_HERE"

if api_key == "INSERT_YOUR_GOOGLE_API_KEY_HERE":
    print("‚ö†Ô∏è NOTE TO JUDGES: Please insert your own Google API Key in line 24 to run this.")
else:
    genai.configure(api_key=api_key)

def gift_agent_final(recipient, budget):
    """
    Main function to run the Gift Scout Agent.
    Args:
        recipient (str): Who the gift is for.
        budget (str): The maximum price.
    Returns:
        str: The AI's suggestions.
    """
    print(f"ü§ñ Gift Scout Agent initializing for: {recipient}...")


    try:
        model = genai.GenerativeModel('gemini-1.5-flash')


        prompt = f"""
        Role: You are an expert Personal Shopper.
        Task: Suggest 3 specific gift ideas for a {recipient}.
        Constraint: Total budget is ${budget}.
        Format: List the item, the estimated price, and why it's a good fit.
        """

        response = model.generate_content(prompt)
        return response.text

    except Exception as e:

        return f"‚ö†Ô∏è API Error (Check Key): {e}"

if __name__ == "__main__":
    # Define test inputs
    test_recipient = "Brother who loves Retro Gaming"
    test_budget = "50"

    print("--- üöÄ STARTING AGENT ---")

    result = gift_agent_final(test_recipient, test_budget)

    print("\n--- üéÅ RECOMMENDATIONS ---")
    print(result)
    print("--------------------------")
```
